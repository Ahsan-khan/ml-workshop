### **Project: Classifying Handwritten Numbers**
- Using Logistic Regression

##### **Step 1**: Loading Required Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Numpy, Pandas, matplotlib, sklearn 
```

##### **Step 2**: Preparing the Data

*Mnist = Modified National Institute of Standards and Technology*

Dataset containing handwritten images of numbers 0-9

```python
#Making a Data Frame of the mnist csv
mnist_df = pd.read_csv("/content/sample_data/mnist_train_small.csv", header=None)
```

```python
mnist_df.head()
```

```python
mnist_df.tail()
```

```python
print("Dataset rows and columns: "+ str(mnist_df.shape))
```

```python
mnist_array = np.array(mnist_df)
mnist_array
```

```python
# Split into features(X) and target(y)
X = mnist_array[:, 1:]
y = mnist_array[:, 0]
```

```python
X[1]
```

```python
#Visualize the image
# Reshaping the array
display_digit = (np.reshape(X[1],(28,28)))
```

```python
# plt is a module we loaded from matplotlib earlier

import matplotlib.pyplot as plt

plt.imshow(display_digit, cmap='gray_r')
```

```python
y[1]
```

```python
#What digits 0-9 look like in our data
fig, ax = plt.subplots(2,5)
for i, ax in enumerate(ax.flatten()):
    im_idx = np.argwhere(y == i)[0]
    plottable_image = np.reshape(X[im_idx], (28, 28))
    ax.imshow(plottable_image, cmap='gray_r')
```

```python
print("Original X data: " + str(len(X)))
print("Original y data: " + str(len(y)))
```

```python
# Reducing the size of our data from 20000 images to 1000 for faster demo purposes
X = X[0:1000]
y = y[0:1000]
```

```python
print("Reduced X data: " + str(len(X)))
print("Reduced y data: " + str(len(y)))
```

```python
#unique pixel values present in an image
# Images are made of a matrix of pixel values (0-255)
np.unique(X)
```

```python
#Checking our label(y) values
# Digits from 0 to 9
np.unique(y)
```

```python
# Splitting data in train and test
from sklearn.model_selection import train_test_split

#test size = 20%
#train size = 80%
#the ML model will never see test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#### **Step 3**: Building ML Model

```python
from sklearn.linear_model import LogisticRegression

#Initializing an instance for a logistic regression model
LR_model = LogisticRegression(max_iter = 10000)
```

#### **Step 4**: Training our Model

```python
#Recall what X_train is
X_train
```

```python
#Recall what y_train is
y_train
```

```python
#Fitting our mnist data onto 
#Only fitting/training on train data not test data

LR_model.fit(X_train, y_train)
```

#### **Step 5**: Testing our Model

```python
#Lets test our model with some images

y_test[10]
```

```python
plt.imshow(X_test[10].reshape(28,28))
```

```python
LR_model.predict(X_test[10].reshape(1,-1))
```

```python
print(f'Accuracy: {LR_model.score(X_test, y_test)}')
```

```python
import openai
openai.api_key = ""

def generate_text(prompt):
    completions = openai.Completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )

    message = completions.choices[0].text
    return message

while True:
    user_input = input("You: ")
    response = generate_text(user_input)
    print("Bot: ", response)
```
